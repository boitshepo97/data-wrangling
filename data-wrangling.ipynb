### 2.Create new columns containing the total score of each of the personality test subscales. To do this, write a function (or functions) that will calculate the total score for each of the subscales , as set out in scoring.

```python
data_rows = personality.values.tolist()

def sum_subscales(data_rows):
    
    all_rows = []
    for row in data_rows:
        all_rows.append(row)
        
    counters_table = []
    add_for_ones = []
    add_for_twos = []
    add_for_threes = []
    add_for_fours = []
    add_for_fives = []
    
    for each_row in all_rows:
        current_row = each_row
        ones = 0
        twos = 0
        threes = 0
        fours = 0
        fives = 0
        
        for each_val in current_row:
            if type(each_val) == str:
                if each_val[1] == '1':
                    ones += int(each_val[4])
                if each_val[1] == '2':
                    twos += int(each_val[4])
                if each_val[1] == '3':
                    threes += int(each_val[4])
                if each_val[1] == '4':
                    fours += int(each_val[4])
                if each_val[1] == '5':
                    fives += int(each_val[4])
                    
        add_for_ones.append(ones)
        add_for_twos.append(twos)
        add_for_threes.append(threes)
        add_for_fours.append(fours)
        add_for_fives.append(fives)
        
    counters_table.append(add_for_ones)
    counters_table.append(add_for_twos)
    counters_table.append(add_for_threes)
    counters_table.append(add_for_fours)
    counters_table.append(add_for_fives)
    return counters_table

# Create rows to work with
counters_table = sum_subscales(data_rows)
df_count = pd.DataFrame(counters_table)
df_count = df_count.transpose()
df_count.shape
df_count = df_count.rename(columns={'0':'Extraversion',
                                    '1':'Agreeableness',
                                    '2':'Conscientiousness',
                                    '3':'Emotional Stability',
                                    '4':'Intellect'})

df_count.columns = ['Extraversion',
                    'Agreeableness',
                    'Conscientiousness',
                    'Emotional Stability',
                    'Intellect']
```

```python
personality = pd.concat([personality, df_count] ,axis=1, ignore_index=False)
```

### 2. Filter the merged data frame so that you get only the applicants who scored less than 30 on emotional stability, conscientiousness AND agreeableness. Print the ID numbers and departments of these applicants to the screen, and also assign these applicants the tag “high_risk” in a new column. All other applicants get the tag “low_risk”

```python
#For high risk recruits.
risk_high = pers_by_dept[(pers_by_dept['Agreeableness'] < 30) & (pers_by_dept['Conscientiousness'] < 30) & (pers_by_dept['Emotional Stability'] < 30)]
filter_high = risk_high[['ID', 'Department']]

#For low risk recruits.
risk_low = pers_by_dept[(pers_by_dept['Agreeableness'] >= 30) | (pers_by_dept['Conscientiousness'] >= 30) | (pers_by_dept['Emotional Stability'] >= 30)]
filter_low = risk_low[['ID', 'Department']]

filter_high['Risk'] = "high risk"
filter_low['Risk'] = "low risk"

filtered_data = pd.concat([filter_high, filter_low])
filtered_data.sort_values(by= 'ID')
```

### 3. Create a new data frame with a count of the number of low and high risk applicants within each department. Let each department be a separate column. 

```python
frequency = pd.DataFrame(filtered_data.groupby('Risk')['Department'].value_counts()).unstack()

#filling all NaN with 0.
frequency = frequency.replace(np.nan, 0)
frequency
```

```python
table = frequency['Department']
table
```

```python
#Changing all values from floats to integers.
table = table.astype(int)
table
```
