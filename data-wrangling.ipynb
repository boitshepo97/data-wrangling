```python
# loading modules
from IPython.display import display, HTML
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
personality = pd.read_csv('personality_scores.csv', sep=';')
```

```python
#checking for duplicates in the data set.
duplicates = personality[personality.duplicated(['ID'])]
print('The duplicate data in this data are:', duplicates, sep='\n')
```

```python
#checking for duplicates in the data set.
personality.drop_duplicates(subset=['ID'], inplace=True)

# Assert Statement for modification.
personality['ID'].unique().sum() == personality['ID'].sum()
```

```python
personality.head()
```

```python
#creating new columns
personality = pd.concat([personality, pd.DataFrame(columns = ['conscientiousness', 'emotional stability', 'openness to new experience', 'agreeableness', 'extraversion'])],sort=True)

personality
```

```python
def sum_subscales(personality, row, number):
    
    count = 0
    for column in personality:
        checks = row[column]
        if row.column[1] == number:
            number = checks[1]
            num = checks[4]
            if number == number:
                count += num
    return count

personality['Extraversion'] = personality.apply(lambda row: sum_subscales(row,'1', personality), axis = 1)
personality
```

```python
departments = pd.read_csv('departments.csv', sep=';')
pers_by_dept = pd.merge(personality, departments).set_index('ID')
```
